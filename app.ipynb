```python
import streamlit as st
import pandas as pd
import numpy as np
import joblib
import json

# ---------------------------------------
# Page Configuration
# ---------------------------------------

st.set_page_config(
    page_title="Predictive Maintenance System",
    page_icon="⚙",
    layout="wide"
)

# ---------------------------------------
# Load Files
# ---------------------------------------

@st.cache_resource
def load_model():
    return joblib.load(
        "Predictive_Maintenance_Model_20260714_152250.joblib"
    )

@st.cache_data
def load_dashboard():
    return pd.read_csv(
        "Dashboard_Data_20260714_152250.csv"
    )

@st.cache_data
def load_config():
    with open(
        "Model_Config_20260714_152250.json",
        "r"
    ) as f:
        return json.load(f)

model = load_model()
dashboard = load_dashboard()
config = load_config()

# ---------------------------------------
# Sidebar
# ---------------------------------------

st.sidebar.title("Navigation")

page = st.sidebar.radio(

    "Select Page",

    [

        "Dashboard",

        "Live Prediction",

        "Batch Prediction",

        "IMS Bearing",

        "Model Information"

    ]

)
```

```python


st.title(
    "Predictive Maintenance and Reliability Early Warning System"
)

st.markdown("---")

st.markdown(
"""
This dashboard predicts machine failure probability and
prioritizes maintenance using an optimized XGBoost model.
"""
)
```

```python
critical = (
    dashboard["Risk Level"]=="Critical"
).sum()

col1,col2,col3,col4=st.columns(4)

col1.metric(
    "Machines",
    len(dashboard)
)

col2.metric(
    "Average Health Risk Index",
    f"{dashboard['Equipment Health Risk Index'].mean():.2f}"
)

col3.metric(
    "Critical Machines",
    critical
)

col4.metric(
    "Decision Threshold",
    f"{config['Threshold']:.2f}"
)

st.markdown("---")
```

```python
st.subheader("Machine Risk Distribution")

risk_count = dashboard["Risk Level"].value_counts()

st.bar_chart(risk_count)
```

```python
st.markdown("---")

st.subheader("Failure Probability Distribution")

st.bar_chart(
    dashboard["Failure Probability"].value_counts(bins=20)
)
```

```python
st.markdown("---")

st.subheader("Top 20 Highest-Risk Machines")

top20 = dashboard.sort_values(
    by="Equipment Health Risk Index",
    ascending=False
).head(20)

st.dataframe(
    top20,
    use_container_width=True
)
```

```python
st.markdown("---")

st.subheader("Machine Records")

risk_option = st.selectbox(
    "Filter by Risk Level",
    ["All"] + list(dashboard["Risk Level"].unique())
)

if risk_option == "All":
    filtered = dashboard
else:
    filtered = dashboard[
        dashboard["Risk Level"] == risk_option
    ]

st.dataframe(
    filtered,
    use_container_width=True
)
```

```python
st.download_button(
    label="Download Predictions",
    data=dashboard.to_csv(index=False),
    file_name="Machine_Predictions.csv",
    mime="text/csv"
)
```
